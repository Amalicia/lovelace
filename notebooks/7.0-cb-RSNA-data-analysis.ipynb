```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("../data/interim/train_labels.csv")
```

```python
plt.figure(figsize=(10,7))
pd.value_counts(df["any"]).plot.bar()
plt.ylabel("Count")
plt.xlabel("Contains Haemorrhage")
plt.show()
print(pd.value_counts(df["any"]))
```

```python
labels = ['epidural', 'intraparenchymal', 'intraventricular', 'subarachnoid', 'subdural']
epidural_count = pd.value_counts(df["epidural"])
intraparenchymal_count = pd.value_counts(df["intraparenchymal"])
intraventricular_count = pd.value_counts(df["intraventricular"])
subarachnoid_count = pd.value_counts(df["subarachnoid"])
subdural_count = pd.value_counts(df["subdural"])

values = [epidural_count[1], intraparenchymal_count[1], intraventricular_count[1], subarachnoid_count[1], subdural_count[1]]
plt.figure(figsize=(10,7))
plt.bar(labels, values)
plt.xticks(rotation='vertical')
plt.ylabel("Count")
plt.xlabel("Haemorrhage Type")
plt.show()

for i in range(len(labels)):
    print("{0} image count: {1}".format(labels[i], values[i]))
```
