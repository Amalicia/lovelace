```python
import pandas as pd
import numpy as np
```

```python
def append_png(image):
    return image + '.png'


raw_data = pd.read_csv('../data/processed/train_labels.csv', index_col=None, dtype=str)
raw_data.fillna('', inplace=True)
raw_data['ImageNo'] = raw_data['ImageNo'].apply(append_png)

raw_data.head(15)
```

```python
labels = set()
for i in range(len(raw_data)):
    labels.update(raw_data['Tags'][i].split(' '))

labels = list(labels)
labels.sort()
labels

labels_dict = {labels[i]:i for i in range(len(labels))}
inv_map = {v:k for k, v in labels_dict.items()}
print(labels_dict)
print(inv_map)
```

```python
def encode(tags, mapping):
    encoding = np.zeros(len(mapping), dtype='uint8')
    tags_list = tags.split(' ')
    for tag in tags_list:
        encoding[mapping[tag]] = 1
    return encoding

raw_data['EncodedTag'] = raw_data.apply(lambda row: encode(row['Tags'], labels_dict), axis=1)
print(raw_data)
```
